```python
import pandas
```

```python
RAW_HOUSING_DATA = pandas.read_csv("1. Regression - Module - (Housing Prices).csv")
```

```python
import matplotlib.pyplot as plt
```

```python
# x variable is default value and y variable represents sale price
# in graph on y axis le6 means the values are like 1000000,2000000,...8000000(not 1,2,3,...8)
plt.plot(RAW_HOUSING_DATA['Sale Price'])
```

```python
# to color the graph in diff color do this;
plt.plot(RAW_HOUSING_DATA['Sale Price'], color = 'orange')
# to label the x axis;
plt.xlabel("Record Number")
# to label the y axis;
plt.ylabel("Sale Price")
# to give a title to our graph;
plt.title("My First Graph")
# to see the changes made;
plt.show()
```

```python
plt.plot(RAW_HOUSING_DATA['Sale Price'], marker = 'o', markerfacecolor='red',markersize='5', color='yellow', linewidth='3',linestyle='dashed')
# the line graph represents the sale price for each record.
# it does not provide us with any specific insight.
```
