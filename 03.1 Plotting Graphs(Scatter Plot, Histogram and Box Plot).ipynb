```python
import pandas
```

```python
RAW_HOUSING_DATA = pandas.read_csv("1. Regression - Module - (Housing Prices).csv")
```

```python
import matplotlib.pyplot as plt
```

```python
# to plot a Scatter Plot;
plt.scatter(x = RAW_HOUSING_DATA['Flat Area (in Sqft)'], y = RAW_HOUSING_DATA['Sale Price'], color = 'orange')
plt.xlabel("Area")
plt.ylabel("Selling Price")
plt.title("Selling Price VS Area")
plt.show()
```

```python
plt.scatter(x = RAW_HOUSING_DATA['No of Bathrooms'], y = RAW_HOUSING_DATA['Sale Price'], color = 'pink')
plt.xlabel("No of Bathrooms")
plt.ylabel("Selling Price")
plt.title("Selling Price VS No of Bathrooms")
plt.show()
```

```python
plt.scatter(x = RAW_HOUSING_DATA['Age of House (in Years)'], y = RAW_HOUSING_DATA['Sale Price'], color = 'cyan')
plt.xlabel("Age pf House (in Years)")
plt.ylabel("Selling Price")
plt.title("Selling Price VS Age of House (in Years)")
plt.show()
```

```python
# How is data is distributed or spread between this age?(like we want to find out the age distribution of houses)
#  --> How many houses are in the range of 3 to 15 years? --> How many houses are in the range of 15 to 25 years ....and so on.
# this helps us to see if the data is spread uniformly,or squed towards lower value, or towards higher value or towards mean value.
# lets plot histogram
plt.hist(RAW_HOUSING_DATA['Age of House (in Years)'], bins = 10)
```

```python
plt.hist(RAW_HOUSING_DATA['Age of House (in Years)'], bins = 10, color='yellow')
plt.xlabel("Age of House(in Years)")
plt.ylabel("No. of Records")
plt.title("Age Wise Distribution of Houses")
plt.show()
```

```python
# for same reasons as of histogram
# lets plot box plot
plt.boxplot(RAW_HOUSING_DATA['Age of House (in Years)'])
```
