```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
transformed_data = pd.read_csv("transformed_housing_data")
```

```python
transformed_data.head()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
# The ANOVA 'ols' function from the statsmodels library does accept spaces in the column names. So, using 'Data['Sale Price']' will generate an error as there is a space between Sale' and 'Price'.
transformed_data = transformed_data.rename(columns = {'Sale Price' : 'Sale_Price'})
```

```python
transformed_data = transformed_data.rename(columns = {'Condition of the House' : 'Condition_of_the_House'})
```

```python
transformed_data = transformed_data.rename(columns = {'Ever Renovated' : 'Ever_Renovated'})
```

```python
transformed_data = transformed_data.rename(columns = {'Waterfront View' : 'Waterfront_View'})
```

```python
transformed_data.head(10)
```

```python
transformed_data['Waterfront_View'].unique()
```

```python
# now lets run ANOVA for sale price and condition of the house
mod = ols('Sale_Price ~ Condition_of_the_House', data = transformed_data).fit()
```

```python
# making one-way ANOVA table
Anova_Table = sm.stats.anova_lm(mod, typ = 2)
```

```python
print(Anova_Table)
# as f value is high and p value is < 0.05, "we accept (Alternative Hypothesis) & reject (Null Hypothesis)"
```

```python
mod = ols('Sale_Price ~ Waterfront_View', data = transformed_data).fit()
sm.stats.anova_lm(mod, typ = 2)
```

```python
mod = ols('Sale_Price ~ Ever_Renovated', data = transformed_data).fit()
sm.stats.anova_lm(mod, typ = 2)
```

```python
mod = ols('Sale_Price ~ Zipcode', data = transformed_data).fit()
sm.stats.anova_lm(mod, typ = 2)
```

```python
# all variables have an impact on target variable(f and p values)
```

# Dummies

```python
# creating dummies
transformed_data = pd.get_dummies(transformed_data, columns = ['Condition_of_the_House'], drop_first = True)
```

```python
transformed_data.head()
```

```python
transformed_data = pd.get_dummies(transformed_data, columns = ['Ever_Renovated','Waterfront_View'], drop_first = True)
```

```python
transformed_data.head()
```

# Binning

```python
Zip_Table = transformed_data.groupby('Zipcode').agg({'Sale_Price' : 'mean'}).sort_values('Sale_Price', ascending = True)
```

```python
Zip_Table.head()
```

```python
# now we need 10 bins with avg sale price & categorize those 71 zipcodes in those 10 bins
Zip_Table['Zipcode_Group'] = pd.cut(Zip_Table['Sale_Price'], bins = 10,
                                   labels = ['Zipcode_Group_0', 'Zipcode_Group_1',
                                             'Zipcode_Group_2', 'Zipcode_Group_3',
                                            'Zipcode_Group_4', 'Zipcode_Group_5',
                                            'Zipcode_Group_6', 'Zipcode_Group_7',
                                            'Zipcode_Group_8', 'Zipcode_Group_9'],
                                   include_lowest = True)
```

```python
Zip_Table = Zip_Table.drop(columns = 'Sale_Price')
```

```python
# now lets join zip table with our dataset
transformed_data = pd.merge(transformed_data, Zip_Table,
                           left_on = 'Zipcode',
                           how = 'left', right_index = True)
```

```python
transformed_data = transformed_data.drop(columns = 'Zipcode') #as we already have grouped this variable as zipcode group.
```

```python
transformed_data.head()
```

```python
# lets now create dummies of this variable
transformed_data = pd.get_dummies(transformed_data, columns = ['Zipcode_Group'], drop_first = True)
```
